```python
plt.scatter(X1[:50], X2[:50], color='red', marker='o', label='setosa') 
plt.scatter(X1[50:100], X2[50:100], color='blue', marker='x', label='versicolor') 
plt.scatter(X1[100:], X2[100:],color='green', marker='+', label='Virginica')
plt.plot(4.8,2.5,color='black',marker='*', label='Sample')
plt.xlabel("Sepal Length (cm)", fontsize=14)
plt.ylabel("Sepal Width (cm)", fontsize=14)
plt.legend(loc=2) 
plt.show()  #sepal length and sepal width 
```

```python
X = iris["data"][:, 3:]
y = (iris["target"] == 2).astype(np.int)
log_reg = LogisticRegression()
log_reg.fit(X, y) 
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], label="Not Iris-Virginica")
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc=2) 
plt.show()
```
